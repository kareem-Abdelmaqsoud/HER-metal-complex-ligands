```python
#how I determined coordination index. The command line prompt starts at 1, not 0 so need to add 1

from rdkit import Chem
from rdkit.Chem import Draw

# Input SMILES string
smiles = "OC(=O)CC(c1ccccc1)(c2ccccc2)c3ccccc3"  

# Convert SMILES to RDKit molecule
mol = Chem.MolFromSmiles(smiles)

# Label atoms with their indices
for atom in mol.GetAtoms():
    atom.SetProp("atomLabel", str(atom.GetIdx()))

# Generate image with atom index labels
img = Draw.MolToImage(mol, size=(300, 300), kekulize=True, wedgeBonds=True)

# Display the image
from IPython.display import display
display(img)
```
